```python
import thinkstats2
import first
import thinkplot
```

```python
live, firsts, others = first.MakeFrames()
```

```python
cdf_firsts = thinkstats2.Cdf(firsts.prglngth)
cdf_others = thinkstats2.Cdf(others.prglngth)
delta = cdf_firsts.Mean() - cdf_others.Mean()
delta
```

```python
sums = [sum(cdf_firsts.Sample(len(firsts))) for _ in range(1000)]
sampling_cdf_firsts = thinkstats2.Cdf(sums)
thinkplot.Cdf(sampling_cdf_firsts)
```

```python
sums = [sum(cdf_others.Sample(len(others))) for _ in range(1000)]
sampling_cdf_others = thinkstats2.Cdf(sums)
thinkplot.Cdf(sampling_cdf_others)
```

```python
cdf1 = sampling_cdf_firsts.Scale(1.0 / len(firsts))
cdf2 = sampling_cdf_others.Scale(1.0 / len(others))
thinkplot.Cdf(cdf2)
```

```python
pmf_diff = cdf1.MakePmf() - cdf2.MakePmf()
```

```python
cdf_diff = pmf_diff.MakeCdf()
thinkplot.Cdf(cdf_diff)
cdf_diff.Mean()
```

```python
sample = cdf_diff.Sample(len(firsts))
thinkstats2.NormalProbabilityPlot(sample)
```

```python
1 - cdf_diff[2 * delta]
```

```python
cdf_diff.ConfidenceInterval(95)
```

```python
mean, std = live.prglngth.mean(), live.prglngth.std()
mean, std
```

```python
import math
mu = mean
sig1 = std / math.sqrt(len(firsts))
sig2 = std / math.sqrt(len(others))
mu, sig1, sig2
```

```python
mu_diff = 0
var_diff = sig1**2 + sig2**2
sig_diff = math.sqrt(var_diff)
var_diff
```

```python
xs, ps = thinkstats2.RenderNormalCdf(mu_diff, sig_diff, -sig_diff*3, sig_diff*3)
thinkplot.Plot(xs, ps)
```

```python
thinkstats2.EvalNormalCdf(-delta, mu_diff, sig_diff)
```

```python
1 - thinkstats2.EvalNormalCdf(delta, mu_diff, sig_diff)
```
